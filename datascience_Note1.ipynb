```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,4])
y=np.array([0,100])
plt.plot(x,y,marker="^")
plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,4,6,7,28,30,26,8,9])
y=np.array([4,8,9,0,2,4,6,7,12])
plt.scatter(x,y,marker="^",color="royalblue")
plt.title("Scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
y=np.array([100,250,500,650,800])
plt.hist(y)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
y=[35,25,25,15]
mylabels=["Train","Bus","Car","Bike"]
plt.pie(y,labels=mylabels)
plt.show()
```
